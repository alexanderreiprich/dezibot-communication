```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(range(0, 10))
y = np.array([100, 23, 44, 12, 55, 85, 69, 33, 75, 2])
plt.title("Plotting 1-D array")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.plot(x, y, color="red", marker="o", label="Array elements")
plt.legend()
plt.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.io as pio

input_file = "C:\\Users\\annad\\Documents\\RU\\putty_singular_data.csv"
dataset = pd.read_csv(input_file, delimiter=";")
df = pd.DataFrame(dataset)
# print(df);
# print(df[df.columns[0]].head())  # Or check for NaNs
# print(df.isnull().sum())
x = df[df.columns[0]]
y = df[df.columns[1]]

fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, mode='lines+markers', name='Array elements', line=dict(color='red')))
fig.update_layout(title="Sensor Data", xaxis_title="time", yaxis_title="amplitude")

pio.renderers.default = 'browser'  # Or 'notebook', 'svg', etc.
fig.show()
```
